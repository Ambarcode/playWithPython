SPLITTING THE FEATURES AND TARGET


```python
#we will predict these 0 by analyzing the other features 
#we'll split the dataset first i.e the target attribute and features

X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

SPLITTING THE DATA INTO TRAINiNG DATA TESTING DATA

```python
X_train , X_test,Y_train , Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape) #checking proportions of the test and train data
```

STANDARDIZING THE DATA

```python
# Scale the data using standardization
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# **MODEL TRAINING**

```python
model=LogisticRegression(max_iter=1000)
```

```python
#now we need to train our machine learning model
model.fit(X_train,Y_train) #this is all about training our model
```

MODEL EVALUATION

```python
#with the help of accuracy score , we can check the accuracy of our model
#Accuracy on Training Data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on Training Data: ",training_data_accuracy)
```

```python
#Accuracy on Test Data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print("Accuracy on Test Data: ",test_data_accuracy)
```
